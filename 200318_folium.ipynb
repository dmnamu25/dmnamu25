```python
import folium
from folium import plugins

import pandas as pd
import time

xlsx = pd.read_excel('Charger_Date.xlsx')

map_osm = folium.Map(location=[33.3616658,126.5204118], zoom_start=10.4)

column = 0
row = 0
a = 0
b = 0

longtitude = []
latitude = []
stat_list = []

points =[]
jeju_time = []


while column + 3 < 147:
    
    i = 0
    
    jeju_time.append(xlsx.iloc[i,column+3])
    
    while i < 205:

        stat = xlsx.iloc[i+1,column+3]
        
        if stat != 0:

            latitude.append(float(xlsx.iloc[i+1, row+1])) 
            longtitude.append(float(xlsx.iloc[i+1, row+2]))

            stat_list.append(stat)
            
        i += 1
        
    while a < len(latitude):
        points.append({'coordinates':[longtitude[a],latitude[a]],'time': jeju_time[b]})
        a += 1

    longtitude = []
    latitude = []
    print(stat_list)
    column += 1
    
    a = 0
    b += 1
    

print(points)

features = [{'type':'Feature','geometry':
             {'type':'Point', 'coordinates':point['coordinates']}
             ,'properties':{'time': point['time']}}  for point in points ]

plugins.TimestampedGeoJson({'type': 'FeatureCollection', 'features': features}, max_speed=40, period='PT1M',duration='PT9M').add_to(map_osm)

plugins.MarkerCluster(features).add_to(map_osm)

map_osm.save('200318.html')
```
